```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = 'https://www.icc-cricket.com/rankings/womens/player-rankings/odi/all-rounder'
```

```python
page = requests.get(url)
page
```

```python
page.content
```

```python
soup = BeautifulSoup(page.content, "html.parser")
print(soup)
```

```python
player=soup.find('td', class_="table-body__cell rankings-table__name name")
player.text
```

```python
team = soup.find('span', class_='table-body__logo-text')
team.text
```

```python
rating = soup.find('td', class_='table-body__cell rating')
rating.text
```

```python
player = []

for i in soup.find_all('td', class_='table-body__cell rankings-table__name name'):
    player.append(i.text.split('\n'))
player
```

```python
team = []
for i in soup.find_all('span', class_="table-body__logo-text"):
    team.append(i.text)
team
```

```python
rating = []
for i in soup.find_all('td', class_="table-body__cell rating"):
    rating.append(i.text)
rating
```

```python
df = pd.DataFrame({'Players':player, 'Team':team, 'Rating':rating})
df
```
